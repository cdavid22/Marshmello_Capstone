# Centroid Calculation for house cordinates.

### **  The housing data comes with features such as no of floors, age of the house  etc for every house in montreal. It also comes with the cordinates of each of the corners of every house listed in the column "geometry". so for example if the plot of the house is a triangle then the corresponding record will have 3 sets of cordinates. the task here is to transform these 3 sets of cordinates into a single set that will be the respresentative of the house. **

```
import pandas as pd
import numpy as np
from shapely.geometry import Point,Polygon
import shapely.wkt

# import geopandas as gpd
```

```
housing = pd.read_csv("uniteevaluationfonciere_df.csv",sep='\t')
housing.head()
```

```
housing.info()
```

```
# Cleaning the geometry column 
housing.geometry = np.where(housing.geometry.str.startswith('P') == False, housing.geometry.str.lstrip('('), housing.geometry)
```

```
# Calculating centroid
centr = [0]*housing.geometry.count()

for i in range(housing.geometry.count()) :
  ref_poly = shapely.wkt.loads(housing.geometry[i])
  centroid = ref_poly.centroid.wkt
  centr[i] = centroid

# housing['centroid'] = centr
```
